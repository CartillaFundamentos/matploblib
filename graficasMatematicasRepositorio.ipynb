# Graficas matematicas con MATPLOTLIB

```python
# importamos la librería y las funciones requeridas
from matplotlib.pyplot import plot, show, xlim, ylim, xticks, legend, figure, subplot
from numpy import linspace, pi, sin, cos
```

```python
# grafica de una lista
y = [1, 2, 5, 3]
plot(y, "*")
show()
```

```python
# grafica de una función
x = [1, 2, 3, 4]
y = [1, 2, 5, 3]
xlim(-5, 10)
ylim(0, 6)
plot(x, y, "*")
show()
```

```python
# grafica de una función trigonométrica

x = linspace(-10*pi, 10*pi, 2560)
y = sin(x)
plot(x, y)
ylim(-2, 2)
show()
```

```python
# grafica de una función trigonométrica

x = linspace(-pi, pi, 256)
y = sin(x)
plot(x, y)
ylim(-2, 2)
xticks([-pi,-pi/2,0,pi/2,pi],[r'$-\pi$',r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$'])
show()
```

```python
# grafica de seno y coseno

x = linspace(-pi, pi, 256)
y = sin(x)
z = cos(x)
plot(x, y, label="seno")
plot(x, z, label="coseno")
legend(loc="upper left")
ylim(-2, 2)
xticks([-pi,-pi/2,0,pi/2,pi],[r'$-\pi$',r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$'])
show()
```

```python
# grafica de seno y coseno - figuras multiples

x = linspace(-pi, pi, 256)
y = sin(x)
z = cos(x)

# figura 1

a = figure(2)
plot(x, y, label="seno")
legend(loc="upper left")

b = figure(20)
plot(x, z, label="coseno")
legend(loc="upper left")

ylim(-2, 2)
xticks([-pi,-pi/2,0,pi/2,pi],[r'$-\pi$',r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$'])
show()
```

```python
# grafica de seno y coseno - subgraficas

x = linspace(-pi, pi, 256)
y = sin(x)
z = cos(x)

# subgraficas

subplot(2,1,1)
plot(x, y, label="seno")
legend(loc="upper left")

subplot(2,1,2)
plot(x, z, label="coseno")
legend(loc="upper left")

ylim(-2, 2)
xticks([-pi,-pi/2,0,pi/2,pi],[r'$-\pi$',r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$'])
show()
```
